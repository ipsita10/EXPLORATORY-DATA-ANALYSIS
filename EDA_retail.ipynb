# **PROFIT AND SALES BY REGION**

```
plt.figure(figsize=[10,15])
df.groupby('Region')['Profit','Sales'].agg(['sum']).plot.bar()
```

# **PROFIT AND SALES BY CATEGORY**

```
plt.figure(figsize=[10,15])
df.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
```

# **PROFIT AND SALES BY SEGMENT**

```
plt.figure(figsize=[10,15])
df.groupby('Segment')['Profit','Sales'].agg(['sum']).plot.bar()
```

# **PROFIT AND SALES BY SHIP MODE**

```
plt.figure(figsize = [10,15])
df.groupby('Ship Mode')['Profit','Sales'].agg(['sum']).plot.bar()
```

# **PROFIT AND SALES BY SUB-CATEGORY**

```
plt.figure(figsize = [10,15])
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
```

1. Here we see that the Tables, Bookcases and Supplies have accounted for losses.

2. Art, Labels, Paper, Envelopes have high profit in relation to sales.



# **DISTRIBUTION OF CATEGORIES**

```
plt.figure(figsize = (7,7))
textprops = {"fontsize":18}
plt.title('Category')
plt.pie(df['Category'].value_counts(), labels = df['Category'].value_counts().index,autopct='%1.1f%%',textprops = textprops)
plt.show()
```

# **DISTRIBUTION OF SUB-CATEGORIES**

```
plt.figure(figsize = (8,8))
textprops = {"fontsize":13}
plt.title('Sub-Category')
plt.pie(df['Sub-Category'].value_counts(), labels=df['Sub-Category'].value_counts().index,textprops = textprops)
plt.show()
```

# **DISTRIBUTION OF DIFFERENT SEGMENTS**

```
plt.figure(figsize = (8,8))
textprops = {"fontsize":13}
plt.title('Segment')
plt.pie(df['Segment'].value_counts(), labels=df['Segment'].value_counts().index,autopct='%1.1f%%',textprops = textprops)
plt.show()
```

```
plt.figure(figsize = (20,6))
plt.subplot(2,1,1)
sns.countplot('Region',hue='Category',data=df)
plt.title('REGIONWISE DISTRIBUTION OF DIFFERENT CATEGORIES')
plt.ylabel('COUNT',fontsize=12)
plt.xlabel('REGION',fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```

plt.figure(figsize = (20,6))
plt.subplot(2,1,2)
sns.countplot('Region',hue='Sub-Category',data=df)
plt.title('REGIONWISE DISTRIBUTION OF SUB-CATEGORIES')
plt.ylabel('COUNT',fontsize=12)
plt.xlabel('REGION',fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

# **PROFIT BY STATE**

```
plt.figure(figsize = (14,12))
df.groupby(['State'])['Profit'].sum().nlargest(50).plot.barh()
```

1. California, New York, Washington and Michigan have accounted for the most profits

2. Texas, Ohio, Pennsylvania, Illinois, North Carolina, Colorado, Tennessey, Arizona, Florida and Oregon have accounted for losses

# **PROFIT BY CITY**

```
plt.figure(figsize=(14,12))
statewise = df.groupby(['City'])['Profit'].sum().nlargest(50)
statewise.plot.barh()
```

# **CONCLUSIONS**
1. Among the Sub-Categories , the sale of Tables, Bookcases and Supplies can be minimized as they accounted for losses. Also Art, Labels, Paper, Envelopes have high profit in relation to sales. So more focus can be put on those items.

2. We can limit sales in states like Texas, Ohio, Pennsylvania, Illinois, North Carolina, Colorado, Tennessey, Arizona, Florida and Oregon, selling where, we incurred losses.

3. More focus can put on selling in states where we gained high profits like New York, California, Washington etc.

4. Furniture has very less profit in relation to sales. So less focus can be put on sale of furniture.

5. More focus should be put on selling Office Supplies and Technology as their profits are high in relation to sales.
